I try to keep five kinds of syntax in my brain at once with Python:

1. keywords/statements
1. functions/methods
1. variables/names
1. punctuation/operators
1. space/indentation


Indentaion in a Jupyter notebook mostly "just works".

Here's an example of all five in use: ```import```, ```for``` and ```in``` are keywords, ```:``` is an operator, ```word``` is a variable (which we both define and use), and ```print()``` is a (built-in) function. And the indentation before ```print``` is required, it won't work if you get rid of it.

````python
import keyword
for word in keyword.kwlist:
    print(word)
````

Voila! That's all the keywords. If I want to learn more about a specific keyword,  I use the "help" function to get some hints, though I don't find it very readable in some cases:

````python
help("while")
````
